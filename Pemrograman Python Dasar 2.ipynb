### Percabangan

Python tidak menggunakan kurung kurawal untuk perulangan dan percabangan seperti yang digunakan C/C++. Python menggunakan `:` dan tab menjorok. Urutan pengecekannya hampir sama seperti C/C++ (`if` -> `elif` (gantinya else if) -> `else`).

Percabangan juga bisa dibuat nested.

```python
Aljabar = 3
Matdis = 4
```

```python
if Aljabar ==5:
  print("Aljabar lumayan bagus")
  if Matdis>5:
    print("Tapi Matdis dah keren")
  else:
    print("Tapi Matdis masih kurang")
elif Aljabar >5:
  print("Keren")
else:
  print("Anda perlu banyak berlatih") 
```

### Perulangan 

Python juga memiliki perulangan seperti C/C++. Cara pertama adalah menggunakan `while`

```python
x:int = 10

while x>1:
  print("Anak ayam turun %d. Mati satu tinggal %d"%(x,x-1))
  x -= 1
  
print("Anak ayam turun %d. Mati satu tinggal induknya"%x)
```

Cara kedua adalah dengan menggunakan loop for. For Python berbeda dengan C/C++. 

Semisal di C/C++, loopnya adalah `for (i=6;i<20;i=i+2) {...}`, di Python kita bisa mengubahnya menjadi seperti di bawah 

```python
for i in range(6,20,2):
  print(i)
```

Iterasi juga bisa dilakukan pada item di list

```python
listbaru = ["Satu","Dua",3]

for i in listbaru:
  print(i)
```

### Fungsi

Definisikan fungsi dengan menggunakan def

```python
def fungsi(x):
    return x**2 +2*x +1
```

```python
print(fungsi(20))
```

Bisa juga dengan type data

```python
def fungsibaru(x:int, kalimat:str,c:float) -> float:
    print(kalimat)
    return x**2 +2*x +1 +c
```

```python
hasil = fungsibaru(20,"Ini adalah kalimat",30)
print(hasil)
```

Urutan argumen bisa ditukar asal namanya dijelaskan

```python
hasil2 = fungsibaru(c=10,kalimat="Ini adalah kalimat",x=50)
print(hasil2)
```

Gunakan tanda bintang `*` untuk argumen yang jumlahnya tidak tetap

```python
def absen(*nama:str):
    print("Mari adik-adik kita absen dulu")
    for i in nama:
        print("Ada %s"%i)
```

```python
absen("Ilham","Kurniawan","Deddy","Corbuzier")
```
