## Numpy

Numpy adalah library Python yang digunakan untuk perhitungan matematis. Hal ini sangat bermanfaat di dunia Data Science

```python
import numpy as np
```

Berbagai fungsi matematis tersedia di Numpy, seperti fungsi trigonometris dan logaritmik. Contohnya, kita diminta mencari

- $cos(65^\circ)$
- $cot(56^\circ)$
- $log_{10}(sin(0.25Ï€))$
- $e^{cos(ln(3.5))}$

```python
a = np.cos(65 * np.pi/180)
b = 1/np.tan(56 * np.pi/180)
c = np.log10(np.sin(0.25*np.pi))
d = np.exp(np.cos(np.log(3.5)))

print(a)
print(b)
print(c)
print(d)
```

Dalam Numpy, kita akan sering bermain dengan tipe data array.
- Array berdimensi satu sering disebut vektor
- Array berdimensi dua sering disebut matriks
- Array berdimensi tiga atau lebih sering disebut tensor (Asal mula kata Tensorflow)

Array mirip `list`, namun hanya bisa memiliki satu tipe data saja

```python
v_a = np.array([1,2,3,4])
v_b = np.array([1,-0.5,1,-0.75])

matriks_a = np.array([[1,2,3,4],
                      [5,6,7,8],
                      [1,3,5,7],
                      [3,4,5,6]])
matriks_b = np.array([[1,0,0,0],
                     [2,3,0,0],
                     [3,4,5,0],
                     [4,5,6,7]])
```

Operasi Perkalian Vektor dan Matriks

```python
#Dot product 2 vektor
print(v_a @ v_b)
print("\n")
#Perkalian Matriks dan vektor
print(matriks_a @ v_a)
print("\n")
# Perkalian matriks dan matriks
print(matriks_a @ matriks_b)
```

Beberapa fungsi aljabar linear Numpy

```python
# Menyelesaikan permasalahan Ax=B
x = np.linalg.solve(matriks_a,v_a)
print(x)
```

```python
# Mencari eigenvalue dan eigenvector matriks
result = np.linalg.eig(matriks_a)

# Daftar eigenvalue
print(result[0])
print("\n")
# Daftar eigenvector
print(result[1])
```
