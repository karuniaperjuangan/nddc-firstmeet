### Variabel dan Tipe Data Python

Python mengenal beberapa tipe data, antara lain

1. Kalimat dalam bentuk `str` (string) 
2. angka dalam bentuk `int` (bil. bulat), `float` (bil. real), atau `complex` (bil.kompleks)
3. Kumpulan data dalam bentuk `list`, `set`, `tuple`
4. `bool` (Benar atau salah)

Dan beberapa jenis lainnya.

Tidak seperti C/C++, kita tidak perlu mendeklarasikan variabel. Cukup set nilai variabel, otomatis tipe datanya diketahui oleh Python.

```python
x = 5
print(type(x))
y = "Lima"
print(type(y))
z = True
print(type(z))
```

Tipe data bisa diubah (casting) bila memungkinkan

```python
a= 5
print(a,type(a))

b= float(a)
print(b,type(b))

c=complex(b)
print(c,type(c))

d= str(c)
print(d,type(d))
```

Tidak semua tipe data bisa dicasting

```python
x = float("Lima")
```

#### Angka dan Operasi Aritmatika

Sesama angka bisa dijumlah Python, tipe data yang berbeda tidak bisa dijumlah

```python
x = 5
y = 6.79
z = 3.01+0.62j
hasil = x+y+z
print(x+y+z)
```

```python
a = 5
b = "enam"
print(a+b)
```

Ada beberapa operasi yang bisa dilakukan dengan angka

```python
a= 5
b = 2

# Penjumlahan
print("Hasil penjumlahan adalah")
print(a+b)

# Pengurangan
print("Hasil pengurangan adalah")
print(a-b)

# Perkalian
print("Hasil perkalian adalah")
print(a*b)

# Pembagian
print("Hasil pembagian adalah")
print(a/b)

# Pembagian menjadi integer (Dalam C/C++ kita bisa samakan dengan hasil = (int) 5/2)
print("Hasil pembagian menjadi integer adalah")
print(a//b)

# Perpangkatan
print("Hasil perpangkatan adalah")
print(a**b)

# Modulus / Sisa Pembagian
print("Sisa pembagian adalah")
print(a%b)
```

Kita bisa meringkas suatu operasi bilangan dengan bilangan itu sendiri (Berlaku untuk semua operator)

```python
# Penjumlahan
x =5
x = x+5
print(x)
#Bisa diringkas menjadi
x = 5
x += 5
print(x)
```

#### String

String adalah tipe data kalimat. String memiliki 2 operator

```python
# Operator Penggabungan

a = "NDDC"
b = "Gabut"
hasil = a+b
print(hasil)

# Operator Perulangan
print(hasil*10)
```

Ada beberapa method yang bisa dijalankan

```python
#Mengecilkan huruf
print(hasil.lower())

#Mengkapitalkan huruf
print(hasil.upper())

#Dan banyak lagi
```

#### Boolean

Boolean adalah tipe data bernilai `True` atau `False` . Kita bisa menggunakan beberapa operasi logika pada boolean

```python
a = True
b = False

print(a and b)
print(a or b)
print(not b)
```

#### List

List adalah kumpulan data yang memiliki urutan, tipe datanya bisa campur aduk, serta ukurannya bisa berubah (tidak seperti Array)

```python
listbaru = [1,"dua",True,5.01+6j]
print(listbaru)
```

Akses datanya mirip pada C/C++, dimulai dari 0

```python
print(listbaru[2])
listbaru[2] = 17
print(listbaru)
```

Kalau di dalam kurung kotak bernilai negatif, berarti kita mengaksesnya dari urutan belakang

```python
print(listbaru[-1]) # Terakhir
```

Kita bisa menggabungkan 2 list

```python
[1,"dua",True,5.01+6j] + [5,4,3]
```

Untuk menambahkan list

```python
listbaru.append(3)
print(listbaru)
```

#### Tuple

Tuple juga merupakan kumpulan data. Hal yang membuatnya beda dengan List adalah kalau sudah dibuat, tidak bisa dimodifikasi.

```python
tuplebaru = (3,4,5,6)
print(tuplebaru)
```

```python
print(tuplebaru[1])
tuplebaru[1] =5
```

#### Set

Set adalah kumpulan data tapi tidak diurutkan. Performanya jauh lebih cepat dari list.

```python
setbaru = {3,2,1,4,"Lima"}
```

```python
print(setbaru)
```

Kita tidak bisa melihat anggota set dari indeksnya

```python
setbaru[3]
```

#### Dictionary

Dictionary adalah tipe data yang berisi kumpulan pasangan `key` dan `value`

```python
dictbaru = {
    "anjing":"segawon",
    "ayam":"pitik",
    "kuda":"jaran",
    "nddc":"gabut"
}
```

```python
print(dictbaru.values)
print(dictbaru.keys)
print(dictbaru["anjing"])
```
